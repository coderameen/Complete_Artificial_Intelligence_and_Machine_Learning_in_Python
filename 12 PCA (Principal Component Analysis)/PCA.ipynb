# PCA(PRINCIPAL COMPONENT ANALYSIS): its a process of figuring out most important feature /principal components that has the most impact on the target variable"

```python
#Its a Technique to reduce dimensionality. suppose we have 3 independent feature we need to reduce 2 independent features by using a new vector space
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
cancer.keys()
```

```python
print(cancer['DESCR'])
```

```python
df = pd.DataFrame(cancer['data'],columns = cancer['feature_names'])
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df)
scaler_data = scaler.transform(df)
scaler_data
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 2)
pca.fit(scaler_data)
x_pca = pca.transform(scaler_data)
scaler_data.shape
```

```python
x_pca.shape
```

```python
#note aboe. the dimension has been reduced from 30 features to 2 features
scaler_data
```

```python
x_pca
```

```python
#WKT, we now have two independent feature, based on target value we will plot Graph
plt.figure(figsize=(8, 6))
plt.scatter(x_pca[:, 0], x_pca[:, 1], c=cancer['target'], cmap='viridis', alpha=0.7)
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.colorbar(label='Target')
plt.show()
```
