```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
#package to ignore warnning
import warnings
warnings.filterwarnings('ignore')
```

```python
#read dataset
dataset = pd.read_csv('Melbourne_housing_FULL.csv')
dataset.head()
```

```python
dataset.shape
```

```python
#considering only useful coulumn
cols_to_use = ['Suburb','Rooms','Type','Method','SellerG','Regionname','Propertycount','Distance','CouncilArea','Bedroom2','Bathroom','Car','Landsize','BuildingArea','Price']
```

```python
cols_to_use
```

```python
dataset = dataset[cols_to_use]
dataset.head()
```

```python
dataset.shape
```

```python
dataset.isna()
```

```python
dataset.isna().sum()
```

```python
cols_to_fill_zero = ['Propertycount','Distance','Bedroom2','Bathroom','Car']
dataset[cols_to_fill_zero] = dataset[cols_to_fill_zero].fillna(0)
```

```python
dataset.isna().sum()
```

```python
dataset['Landsize'] = dataset['Landsize'].fillna(dataset.Landsize.mean())
dataset['BuildingArea'] = dataset['BuildingArea'].fillna(dataset.BuildingArea.mean())
dataset.isna().sum()
```

```python
dataset.dropna(inplace=True)
dataset.isna().sum()
```

```python
#Now, NAN values are removed, but we can see labels 
```

```python
dataset.head()
```

```python
dataset = pd.get_dummies(dataset,drop_first=True)
dataset.head()
```

```python
#create x and y axis
x = dataset.drop('Price',axis=1)
x.head()
```

```python
y = dataset['Price']
y.head()
```

```python
#train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3, random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
#score: Testing
model.score(x_test,y_test)#13%
```

```python
#score: training
model.score(x_train,y_train)#68%
```

```python
#as above, in Train Performance is high and Testing Performance is Low => Overfitting
```
