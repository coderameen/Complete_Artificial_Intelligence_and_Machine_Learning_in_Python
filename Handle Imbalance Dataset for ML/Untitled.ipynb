### Imbalance Dataset

##### A dataset with an unequal class distribution

```python
import pandas as pd
import numpy as np
```

```python
#load dataset
df = pd.read_csv('creditcard.csv')
df.head()
```

```python
df['Class'].value_counts()
```

```python
#Separrating the legit and fraudulant transactions
legit = df[df.Class == 0]
fraud = df[df.Class == 1]
print(legit.shape, fraud.shape)
```

```python
#idea: create a new sample in which it contains similar distribution of legit and fraudulent transaction
legit_sample = legit.sample(n=492)
legit_sample.shape
```

```python
#concatinate the two dataframes(ie. legit_sample,fraud)
new_dataset = pd.concat([legit_sample,fraud],axis=0)
new_dataset.head()
```
